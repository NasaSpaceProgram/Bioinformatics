```python
from math import exp
def Partition_of_QRotor(n):
    oupt = []
    for j in n:
        count = 0
        for i in range(j+1):
            try:
                count += (2*i+1)*exp(-i*(i+1))
            except:
                print(i)
        oupt.append(count)
    return(oupt)

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation


fig = plt.figure()
ax = plt.axes(xlim=(0, 100), ylim=(0, 2))
x = range(100)
y=Partition_of_QRotor(x)
plt.title("Z vs j_max")
plt.ylabel("Z")
plt.xlabel("j_max")
plt.plot(x,y)
fig.set_size_inches(18.5, 10.5)
fig.savefig('Z_vs_j_max.png', dpi=100)
```
