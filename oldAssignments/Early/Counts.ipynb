```python
from Bio import SeqIO
record = SeqIO.read("lac_operon.fasta", "fasta")
Description = record.description
#print(Description)
Sequence = str(record.seq)
print(Sequence)
```

```python
lis = ["A","T","C","G"]
dinuc = []
for i in lis:
    for j in lis:
        dinuc.append(i + j)
print(dinuc)
```

```python
dic = {}
counts = []
for nuc in dinuc:
    count = Sequence.count(nuc)
    dic[nuc] = count
    counts.append(count)
    
N = len(Sequence)
for nuc in dinuc:
    #percent = round((dic[nuc]*100/N),2)
    print(nuc +": Count: " + str(dic[nuc]))
    
    
    
    
    
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt

sequence = dinuc
y_pos = np.arange(len(sequence))

plt.bar(y_pos, counts, align='center', alpha=0.5)
plt.xticks(y_pos, sequence)
plt.ylabel('Number of bases')
plt.title('Base frequency histogram')

plt.show()
```
